```python
# SIMPLE LINEAR REGRESSION

# 1.IMPORTING LIBRARY


import pandas as pd

# "import pandas" - refers importing all functions under pandas and it will be loaded in a particular memory.

# pandas is a library function,it can handle all types of tables.It can read the file which is related to the database.

# It has multiple functions and also globalized for the users.

# as pd- means converting or assigning the word pandas as pd in simple.

# This will be saved once it run after execution.
```

```python
# DATA COLLECTION PHASE

dataset=pd.read_csv("Salary_Data.csv")

# pd=class

# read csv=function

# pd.read_csv refers - The function read_csv is called from the class pd.

# The dataset file is passed within the bracket with double quotes("Salary_Data.csv")

# The file name sholud be same as mentioned in the database otherwise File not Found Error will occur.

# The prticular dataset file should be under the same folder, here under SLR otherwise File not Found Error will occur.
```

```python
dataset

# To view datas or content in the dataset file, or can display using print statement.

# displays all the column and rows as input and output.

# In this salary dataset YearsExperience column is the input and Salary column is the output.

# For this dataset preprocessing is not required becoz it is preprocessed data. 
```

```python
#INPUT AND OUTPUT SPLIT UP PHASE

#INPUT SPLIT

independent=dataset[["YearsExperience"]]

# Assigning independent variable, also it is input.
# Input column ie., YearsExperience from the table is seperated or extracted or fetchedd by command dataset[["column name"]] ie.,YearsExperience
# To fetch more than one column use comma follow by column name.
# Column name should be same as mentioned in the dataset without spelling mistake. 
```

```python
independent

# To print to check the content present in the independent variable or input column.
```

```python
# OUTPUT SPLIT

dependent=dataset[["Salary"]]

# Similarly the dependent refers output column, it is also seperately fetched from the dataset table by aassigned 
# The column name should be same as such in dataset
```

```python
dependent

# To view the content or datas available in the output column
# Both the i/p and o/p datas are seperated from the table
```

```python
# TRAINING AND TEST SET SPLIT UP

from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test= train_test_split(independent,dependent,test_size=0.30,random_state=0)

# sklearn is a library function.
# sklearn supports Machine Learning.
# it is importing only train and test function from the folder modelselection under sklearn.
# Only required functions are called rather all functions like pandas, therfore memory consuption will be less.

# X_train=training set input
# X_test =test set input
# y_train= training set output
# y_test = test set output

# the size of training and test set will be in the ratio 60:40,70:30,80:20
# underscore is used in middle of the word is to join and to make single word also to understand easily.
# For the function train_test_split the parameters are passed inside the bracket as independent,dependent with test size and random state

# X means input and Y means output
# 4 variables  created with the position, 1st- position is meant for X_train input
#                                         2nd-                       X_test input
#                                         3rd-                       y_train output
#                                         4th-                       y_test output
```

```python
# LEARNING PHASE

# Ml algorithms are provided by sklearn library
# LinearRegression is called or imported from the function linear_model
# Spelling should be checked to avoid errors

# regressor=LinearRegression() means the function under LinearRegression is assigned to the variable regressor
#regressor.fit() means calling thr fit function through access operator
# fit(X_train,Y_train) means the value should given as input or the values to substitute
# SLR- y=wx+b where, weight & input + bais
# model is created by fit method
# LinearRegression()
# created model part is executed ie model is calculated

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
weight=regressor.coef_
weight

# The model value is checked by variable "weight=regressor.coef_" means the coef_ values of slope and bais will be saved in the variable weight
```

```python
bais=regressor.intercept_

#To know the bais value
bais
```

```python
# EVALUATION 

# The model,weight,bais values are saved in the variable regressor
# To evaluate the model test dataset is used ie., X_test
# creating Y_pred variable and evaluation made for testdata ie., x_test

Y_pred=regressor.predict(X_test)
```

```python
# function for finding r score value
# r_score variable is created and for same funcion input is given as actual(y_test) and predicted value(Y_pred)

from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)

# After running r score the result will be between 0-1, the model performance is based on the value if result is near to 1 meaning its a good model. 
```

```python
r_score
```

```python
# TO SAVE MODEL

#pickle is the funtion to save the model
# assign filename as finalized_model_linear.sav
# .sav is the extention filename
# wb means write binary, rd means read binary

import pickle
filename="finalized_model_linear.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
# TO LOAD THE MODEL FOR CHECKING

loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[13]])

# loaded_model is the variable name created
# pickle.load is the function to load. rb means read
```
